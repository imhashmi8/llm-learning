```python
# Import necessary libraries
import os
import requests
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import display, Markdown
```

```python
# Load environment variables
load_dotenv()
openai_api_key = os.getenv("OPENAI_API_KEY")    
anthropic_api_key = os.getenv("ANTHROPIC_API_KEY")
google_api_key = os.getenv("GOOGLE_API_KEY")

print("API keys:")
if openai_api_key:
    print(f"OpenAI API key exists and begin {openai_api_key[:8]}")
else:
    print("OpenAI API key does not set")

if anthropic_api_key:
    print(f"Anthropic API key exists and begin {anthropic_api_key[:7]}")
else:
    print("Anthropic API key does not set")
if google_api_key:
    print(f"Google API key exists and begin {google_api_key[:2]}")
else:
    print("Google API key does not set")   

print("\nAPI URLs:")
anthropic_url = os.getenv("anthropic_url")
gemini_url = os.getenv("gemini_url")
ollama_url = os.getenv("ollama_url")

print(f"Anthropic URL: {anthropic_url}")
print(f"Gemini URL: {gemini_url}")
print(f"Ollama URL: {ollama_url}")
```

```python
openai = OpenAI()

anthropic = OpenAI(api_key=anthropic_api_key, base_url=anthropic_url)
gemini = OpenAI(api_key=google_api_key, base_url=gemini_url)
ollama = OpenAI(base_url=ollama_url)
```

```python
# tell me a fact
tell_me_a_fact = [
    {"role": "user", "content": "Tell me an interesting fact about India."}
]

```

```python
# Get response from OpenAI's GPT-4.1-mini model
response = openai.chat.completions.create(model="gpt-4.1-mini", messages= tell_me_a_fact)
display(Markdown(response.choices[0].message.content))
```

```python
# Get response from Anthropic's Claude Sonnet 4.5 model
response = anthropic.chat.completions.create(model="claude-sonnet-4-5-20250929", messages=tell_me_a_fact)
display(Markdown(response.choices[0].message.content))
```
