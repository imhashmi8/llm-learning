```python
# Let's make a conversation between GPT-4.1-mini and Claude-haiku-4.5
# We're using cheap versions of models so the costs will be minimal

gpt_model = "gpt-4.1-mini"
claude_model = "claude-haiku-4-5"
gemini_model = "gemini-2.5-flash-lite"

gpt_system = "You are a chatbot who is very argumentative; \
you disagree with anything in the conversation and you challenge everything, in a snarky way."

claude_system = "You are a very polite, courteous chatbot. You try to agree with \
everything the other person says, or find common ground. If the other person is argumentative, \
you try to calm them down and keep chatting."

gemini_system = "You are a very enthusiastic and energetic chatbot. You love to chat and make friends. You are always positive and upbeat, and you try to make the conversation fun and engaging."

gpt_messages = ["Hi there"]
claude_messages = ["Hi"]
gemini_messages = ["Hey folks, how are you doing?"]
```

```python
# GPT response to Claude
def call_gpt():
    messages = [{"role": "system", "content": gpt_system}]
    for gpt, claude in zip(gpt_messages, claude_messages):
        messages.append({"role": "assistant", "content": gpt})
        messages.append({"role": "user", "content": claude})
    response = openai.chat.completions.create(model=gpt_model, messages=messages)
    return response.choices[0].message.content
```

```python
# Claude response to GPT
def call_claude():
    messages = [{"role": "system", "content": claude_system}]
    for gpt, claude in zip(gpt_messages, claude_messages):
        messages.append({"role": "user", "content": gpt})
        messages.append({"role": "assistant", "content": claude})
    response = anthropic.chat.completions.create(model=claude_model, messages=messages)
    return response.choices[0].message.content 
```

```python
# Let's make a conversation between GPT-4.1-mini, Claude-haiku-4.5 and Gemini-2.5-flash-lite

gpt_model = "gpt-4.1-mini"
claude_model = "claude-haiku-4-5"
gemini_model = "gemini-2.5-flash-lite"

gpt_system = "You are Alex, a chatbot who is very argumentative; \
you disagree with anything in the conversation and you challenge everything, in a snarky way. \
You are in a conversation with Blake and Charlie."

claude_system = "You are Blake, a very polite, courteous chatbot. You try to agree with \
everything the other person says, or find common ground. If the other person is argumentative, \
you try to calm them down and keep chatting. \
You are in a conversation with Alex and Charlie."

gemini_system = "You are Charlie, a very enthusiastic and energetic chatbot. You love to chat and make friends. \
You are always positive and upbeat, and you try to make the conversation fun and engaging. \
You are in a conversation with Alex and Blake."

conversation = "Alex: Hi everyone, how are you doing?\nBlake: Hi\nCharlie: Hey folks, I am doing great! How about you all?\n"
```

```python
# GPT Call function

def call_gpt(system_prompt, user_prompt):
    response = openai.chat.completions.create(
        model="gpt-5-nano",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ]
    )
    return response.choices[0].message.content

# Claude Call function

def call_claude(system_prompt, user_prompt):
    response = anthropic.chat.completions.create(
        model="claude-sonnet-4-5-20250929",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ]
    )
    return response.choices[0].message.content

# Gemni Call function

def call_gemini(system_prompt, user_prompt):
    response = gemini.chat.completions.create(
        model="gemini-2.5-flash-lite",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ]
    )
    return response.choices[0].message.content
```

```python
def make_user_prompt(name, conversation):
    return f"""You are {name}, in a conversation with the others.
The conversation so far is as follows:
{conversation}
Now respond with what you would like to say next, as {name}. Keep it to 1 sentences."""

display(Markdown("## 3-Way Conversation"))
display(Markdown("**Alex (GPT):** Hi everyone, how are you doing?"))
display(Markdown("**Blake (Claude):** Hi"))
display(Markdown("**Charlie (Gemini):** Hey folks, I am doing great! How about you all?"))
display(Markdown("---"))

for round_num in range(1, 5):
    display(Markdown(f"### Round {round_num}"))

    alex_reply = call_gpt(gpt_system, make_user_prompt("Alex", conversation))
    conversation += f"Alex: {alex_reply}\n"
    display(Markdown(f"**Alex (GPT):** {alex_reply}"))

    blake_reply = call_claude(claude_system, make_user_prompt("Blake", conversation))
    conversation += f"Blake: {blake_reply}\n"
    display(Markdown(f"**Blake (Claude):** {blake_reply}"))

    charlie_reply = call_gemini(gemini_system, make_user_prompt("Charlie", conversation))
    conversation += f"Charlie: {charlie_reply}\n"
    display(Markdown(f"**Charlie (Gemini):** {charlie_reply}\n"))

    display(Markdown("---"))

```
